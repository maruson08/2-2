# **5. 배열의 합치기 기능과 형태 모양 변경이 가능하다.**

#위아래/가로세로로 합치기
axis → 어느 축을 따라 합칠지 지정

axis=0 : 세로 방향(행 방향, 위아래)

axis=1 : 가로 방향(열 방향, 좌우)

axis를 생략하면 기본값이 0 → 위아래 합치기

```python
#합치기
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
array3 = np.concatenate([array1, array2])

print(array3.shape)
print(array3)
```

```python
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

# axis=0 : 세로(위아래)
result0 = np.concatenate([array1, array2], axis=0)
print(result0)
```

```python
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

result1 = np.concatenate([array1, array2], axis=1)
print(result1)
```

```python
#모양 바꾸기
array1 = np.array([1, 2, 3, 4])
array2 = array1.reshape((2, 2))
print(array2.shape)
print(array2)
```

```python
#형태 나누기
array = np.arange(8).reshape(2, 4)
print(array)

left, right = np.split(array, [2], axis=1) #열기준 2에서 나눠라.

print(left.shape)
print(right.shape)

print(left)
print(right)
print(right[1][1])
```

```python
array = np.arange(8).reshape(2, 4)
left, right = np.split(array, [1], axis=0) #행기준 인덱스1에서 나눠라.

print(left.shape)
print(right.shape)
print(right[0][1])
```

```python
# Numpy 원소의 값을 조건에 따라 바꿀 때
# 반복문을 이용할 때보다 매우 빠르게 동작
# 대체로 이미지 처리(Image Processing)에서 자주 활용
array1 = np.arange(16).reshape(4, 4)
print(array1)

array2 = array1 < 10
print(array2)

array1[array2] = 100
print(array1)
```

```python
#다양한 집계함수

array = np.arange(16).reshape(4, 4)
print(array)
print("최대값:", np.max(array), array.max(axis=1), array.max(axis=0))
print("최소값:", np.min(array))
print("합계:", np.sum(array))
print("평균값:", np.mean(array))
```

#브로드캐스팅


*   서로 다른 모양(shape)의 배열(array)끼리 산술 연산(+, -, *, / 등)을 수행할 때 자동으로 모양을 맞춰 연산이 가능하도록 확장하는 기능
*  작은 배열을 큰 배열의 크기에 맞춰 반복해서 계산


<규칙>
1.  배열의 차원 수가 다르면, 작은 배열의 앞쪽(왼쪽)에 1을 추가해서 차원을 맞춤

2.  각 차원의 크기가 같거나 하나가 1이면 브로드캐스팅 가능
3.  크기가 같지 않고, 1도 아니면 오류









```python
import numpy as np

a = np.array([1, 2, 3])
b = 2

print(a + b)  #스칼라를 배열 크기만큼 자동으로 늘려서 계산, b가 [2, 2, 2]로 늘려짐
```

```python
a = np.array([[1, 2, 3],   #shape: (2, 3)
              [4, 5, 6]])
b = np.array([10, 20, 30])  #shape: (1, 3)

print(a + b)
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])  #(2, 3)
b = np.array([[10],
              [20]])  #(2, 1)

print(a + b)
```

```python
#1차원 벡터 내적

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_product = np.dot(a, b)
print(dot_product)  # 32
```

```python
A = np.array([[1,2],[3,4]])
print(A)

B = np.array([[5,6],[7,8]])
print(B)

print(A @ B)
print(np.dot(A, B))

print(A)
print(A.T)

```

# **7. numpy 실제 활용**


*  이미지 표현
* 머신러닝
*   벡터 연산, 물리학등의 계산에서 활용


```python
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# 1. 이미지 생성
# ===============================
# 5x5 흑백 이미지 (0~255) 각 숫자는 픽셀 밝기 값 (0: 검은색, 255: 흰색)
original_image = np.array([[50, 80, 90, 100, 120],
                           [60, 90, 110, 130, 150],
                           [70, 100, 120, 140, 160],
                           [80, 110, 130, 150, 170],
                           [90, 120, 140, 160, 180]])

# 밝기 증가 (50 더하기, 최대 255) np.clip 사용 → 255를 초과하지 않도록 제한
brighter_image = np.clip(original_image + 50, 0, 255)

# 컬러 이미지 예시 (RGB 5x5x3)
color_image = np.random.randint(0, 256, (5,5,3))

# ===============================
# 2. 이미지 화면 출력
# ===============================
fig, axes = plt.subplots(2, 2, figsize=(8,8))        #여러 그래프(서브플롯)를 한 번에 만들 때 사용

# 원본 흑백
axes[0,0].imshow(original_image, cmap='gray')
axes[0,0].set_title("Original Grayscale")
axes[0,0].axis('off')

# 밝기 증가 흑백
axes[0,1].imshow(brighter_image, cmap='gray')
axes[0,1].set_title("Brighter Grayscale")
axes[0,1].axis('off')

# 컬러 이미지
axes[1,0].imshow(color_image)
axes[1,0].set_title("Color Image")
axes[1,0].axis('off')

# 컬러 이미지 흑백 변환
gray_from_color = np.dot(color_image[...,:3], [0.2989, 0.5870, 0.1140])
axes[1,1].imshow(gray_from_color, cmap='gray')
axes[1,1].set_title("Color → Grayscale")
axes[1,1].axis('off')

plt.tight_layout()
plt.show()
```
